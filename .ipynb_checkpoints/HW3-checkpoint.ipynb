```python
#Setup conections to repo
import geopandas as gpd
url = "https://github.com/meade68/DACSS_690D_Week13_HW3/raw/refs/heads/main/DATA/tl_2020_19_tract.zip"
IA_TRACTS = gpd.read_file(url)
url = "https://github.com/meade68/DACSS_690D_Week13_HW3/raw/refs/heads/main/DATA/ACSDT5Y2020.B25002_2025-12-04T211012/ACSDT5Y2020.B25002-Data.csv"
IA_DATA= gpd.read_file(url)
```

```python
IA_TRACTS.plot()
IA_TRACTS.head()
```

```python
#Drop original column  names. Rename the column names
IA_DATA = IA_DATA.iloc[1:].reset_index(drop=True)
IA_DATA = IA_DATA.iloc[:, :8]
IA_DATA.columns = ["Geography","Geographic Area Name","Estimate!!Total:",	"Margin of Error!!Total:",	"Estimate!!Total:!!Occupied",	"Margin of Error!!Total:!!Occupied","Estimate!!Total:!!Vacant","Margin of Error!!Total:!!Vacant"]
IA_DATA.head()
```

```python
#Trim/replace the first part of the string 
IA_DATA["Geography"] = IA_DATA["Geography"].str.replace("1400000US", "", regex=False)
IA_DATA.head()
```

```python
IA_MERGED = IA_TRACTS.merge(IA_DATA, left_on="GEOID", right_on="Geography", how="inner")
IA_MERGED.plot()
IA_MERGED.head()
```

```python
#calculate percentage vacant
IA_MERGED["PCT_VACANT"]=IA_MERGED["Estimate!!Total:!!Vacant"].astype(int) / IA_MERGED["Estimate!!Total:"].astype(int)
```

```python
IA_MERGED.plot(
    "PCT_VACANT",
    scheme="quantiles",
    cmap="PuBuGn",
    legend=True,figsize=(12, 10))
```

```python
from libpysal.graph import Graph
from numpy.random import seed
seed(1)
IA_MERGED_perim=Graph.build_contiguity(IA_MERGED,by_perimeter=True)
```

```python
#build matrix
IA_MERGED_perim=IA_MERGED_perim.transform("r")
```

```python
# get spatial lag of PCT_VACANT
ylag = IA_MERGED_perim.lag(IA_MERGED["PCT_VACANT"])
```

```python
IA_MERGED=IA_MERGED.assign(PCT_VACANT_lagged=ylag)
IA_MERGED.head()
```

```python
IA_MERGED.plot.scatter("PCT_VACANT","PCT_VACANT_lagged")
```
