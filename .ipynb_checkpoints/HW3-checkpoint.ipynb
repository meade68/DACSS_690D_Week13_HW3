```python
#Setup conections to repo
import geopandas as gpd
url = "https://github.com/meade68/DACSS_690D_Week13_HW3/raw/refs/heads/main/DATA/tl_2020_19_bg.zip"
IA_TRACTS = gpd.read_file(url)
url = "https://github.com/meade68/DACSS_690D_Week13_HW3/raw/refs/heads/main/DATA/DECENNIALDHC2020.H3_2025-12-10T220427/DECENNIALDHC2020.H3-Data.csv"
IA_DATA= gpd.read_file(url)
```

```python
IA_TRACTS.plot()
IA_TRACTS.head()
```

```python
IA_DATA.head()
```

```python
#Drop original column  names. Rename the column names
IA_DATA = IA_DATA.iloc[1:].reset_index(drop=True)
IA_DATA = IA_DATA.iloc[:, :5]
IA_DATA.columns = ["Geography","Geographic Area Name",	"Total",	"Occupied",	"Vacant"]
IA_DATA.columns
```

```python
#Trim/replace the first part of the string 
IA_DATA["Geography"] = IA_DATA["Geography"].str.replace("1500000US", "", regex=False)
IA_DATA.head()
```

```python
IA_MERGED = IA_TRACTS.merge(IA_DATA, left_on="GEOID", right_on="Geography", how="inner")
IA_MERGED.plot()
IA_MERGED.head()
```

```python

#calculate percentage vacant
import numpy as np
import pandas as pd

IA_MERGED["Vacant"] = pd.to_numeric(IA_MERGED["Vacant"], errors="coerce")
IA_MERGED["Total"]  = pd.to_numeric(IA_MERGED["Total"], errors="coerce")

IA_MERGED["PCT_VACANT"] = IA_MERGED["Vacant"] / IA_MERGED["Total"] * 100

IA_MERGED["PCT_VACANT"] = IA_MERGED["PCT_VACANT"].fillna(0)
```

```python
IA_MERGED.head()
```

```python
IA_MERGED.plot(
    "PCT_VACANT",
    scheme="quantiles",
    cmap="PuBuGn",
    legend=True,figsize=(12, 10))
```

```python
from libpysal.graph import Graph
from numpy.random import seed
seed(1)
IA_MERGED_perim=Graph.build_contiguity(IA_MERGED,by_perimeter=True)
```

```python
#build matrix
IA_MERGED_perim=IA_MERGED_perim.transform("r")
```

```python
# get spatial lag of PCT_VACANT
ylag = IA_MERGED_perim.lag(IA_MERGED["PCT_VACANT"])
```

```python
IA_MERGED=IA_MERGED.assign(PCT_VACANT_lagged=ylag)
IA_MERGED.head()
```

```python
IA_MERGED.plot.scatter("PCT_VACANT","PCT_VACANT_lagged")
```

OOf.

```python
IA_MERGED.plot(
    "PCT_VACANT_lagged",
    scheme="quantiles",
    cmap="PuBuGn",
    legend=True,figsize=(12, 10))
```

```python
import esda # from pysal

MoranGlobal_Housing = esda.Moran(IA_MERGED['PCT_VACANT'], IA_MERGED_perim)
MoranGlobal_Housing.I,MoranGlobal_Housing.p_sim
```

```python
lisa = esda.Moran_Local(IA_MERGED['PCT_VACANT'], IA_MERGED_perim)
lisa.plot(IA_MERGED,crit_value=0.05,figsize=(10,13),legend=True)
```

```python
# get the quadrants
IA_MERGED['PCT_VACANT_lisa'] = lisa.get_cluster_labels(crit_value=0.05)
# see the count
IA_MERGED['PCT_VACANT_lisa'].value_counts()
```

```python
TheStats=["mean", "min", "max","var"]


# stats by the lisa cluster
grouped_stats = IA_MERGED.groupby('PCT_VACANT_lisa')['PCT_VACANT'].agg(TheStats)

# stats for the entire, ungrouped column
global_stats = IA_MERGED['PCT_VACANT'].agg(TheStats)

# Convert 'global_stats' (Series) to a DataFrame with an appropriate index name
global_stats_df = pd.DataFrame(global_stats).T
global_stats_df.index = ['Global/Total'] # simple but key

# Combine both
pd.concat([grouped_stats, global_stats_df])
```

```python
##knowing
IA_MERGED['PCT_VACANT_lisa'].unique()
```

```python
oldLabels=['Insignificant', 'Low-Low', 'High-High', 'High-Low', 'Low-High']

# 3 will be at the middle
newLabels = ['3 no_pattern', '4 coldSpot','2 hotSpot','1 hotOutlier' , '5 coldOutlier']

labels = dict(zip(oldLabels, newLabels))

labels
```

```python
IA_MERGED.replace({'PCT_VACANT_lisa':labels},inplace=True)

## see the count
IA_MERGED['PCT_VACANT_lisa'].value_counts().sort_index()
```

```python
##custom colors- respect the previous order

import matplotlib.pyplot as plt
myColMap = plt.get_cmap('PuOr', 5)


##plot the map

IA_MERGED.plot(column='PCT_VACANT_lisa',
                categorical=True,
                cmap=myColMap,
                linewidth=0.1,
                edgecolor='k',
                legend=True,
                legend_kwds={'bbox_to_anchor': (0.3, 0.3)},
                figsize=(12,12))
```

```python
lisa = esda.Moran_Local(IA_MERGED['PCT_VACANT_lagged'], IA_MERGED_perim)

# get the quadrants
IA_MERGED['PCT_VACANT_lagged_lisa'] = lisa.get_cluster_labels(crit_value=0.05)
# see the count
IA_MERGED['PCT_VACANT_lagged_lisa'].value_counts()

TheStats=["mean", "min", "max","var"]


# stats by the lisa cluster
grouped_stats = IA_MERGED.groupby('PCT_VACANT_lagged_lisa')['PCT_VACANT_lagged'].agg(TheStats)

# stats for the entire, ungrouped column
global_stats = IA_MERGED['PCT_VACANT_lagged'].agg(TheStats)

# Convert 'global_stats' (Series) to a DataFrame with an appropriate index name
global_stats_df = pd.DataFrame(global_stats).T
global_stats_df.index = ['Global/Total'] # simple but key

# Combine both
pd.concat([grouped_stats, global_stats_df])


IA_MERGED.replace({'PCT_VACANT_lagged_lisa':labels},inplace=True)

## see the count
IA_MERGED['PCT_VACANT_lagged_lisa'].value_counts().sort_index()

##custom colors- respect the previous order

import matplotlib.pyplot as plt
myColMap = plt.get_cmap('PuOr', 5)


##plot the map

myColMap = plt.get_cmap('PuOr', 5)


##plot the map

IA_MERGED.plot(column='PCT_VACANT_lagged_lisa',
                categorical=True,
                cmap=myColMap,
                linewidth=0.1,
                edgecolor='k',
                legend=True,
                legend_kwds={'bbox_to_anchor': (0.3, 0.3)},
                figsize=(12,12))
```
